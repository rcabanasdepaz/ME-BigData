# Ejemplo de HMM: Dado Trucado

## Configuración

En primer lugar, instalamos el paquete que implementa la funcionalidad sobre HMMs.

```r
install.packages("HMM")
library("HMM")
```

## Modelo y datos

```r

Z_dom = c("Ok", "Cargado")
X_dom = 1:6


# P(Z_t | Z_{t-1})
probTransicion = matrix(c(0.95, 0.05, 0.1, 0.90), c(length(Z_dom),
                                                 length(Z_dom)), byrow = T)

# P(X_t | Z_t)
probEmision = matrix(c(rep(1/6, 6), c(rep(0.1, 5), 0.5)),
                       c(length(Z_dom), length(X_dom)), byrow = TRUE)



# Construye el modelo
hmm = initHMM(Z_dom, X_dom, transProbs = probTransicion, emissionProbs = probEmision)
hmm
```

```r
# Genera un conjunto de datos de forma aleatoria
ndatos = 200
datos = simHMM(hmm, ndatos)
datos
```

## Inferencia

```r
f = forward(hmm, datos$observation)
b = backward(hmm, datos$observation)
```

```r
# Filtrado
pX1_k = exp(f["Ok",]) + exp(f["Cargado",])
filtrado = exp(f)
filtrado["Ok",] = filtrado["Ok",] / pX1_k
filtrado["Cargado",] = filtrado["Cargado",] / pX1_k
filtrado
```

```r
# Suavizado
suavizado = exp((f + b)) / pX1_k[ndatos]
suavizado
```

```r
options(repr.plot.width=20, repr.plot.height=10)

plot(suavizado[2,], type="l",ylim = c(-0.1, 1), pch = 3,
     xlab = "Tirada", ylab="Prob(cargado)", bty = "n", yaxt = "n")
axis(2, at = c(0,0.5,1))
points(filtrado[2,],type="l",col="red")
for (i in 1:ndatos) {
  if (datos$states[i] == "Ok")
    rect(i, -0.1, i + 1, 0, col = "green", border = NA)
  else rect(i, -0.1, i + 1, 0, col = "red", border = NA)
}
```
