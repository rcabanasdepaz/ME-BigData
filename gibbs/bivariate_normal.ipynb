# Ejemplo Muestreo de Gibbs con una Bivariada Normal

```r

gibbs_bivariate_normal = function(theta0, n_points, rho) {
  theta = matrix(theta0, nrow=n_points, ncol=2, byrow=TRUE)
  v = sqrt(1-rho^2)
  for (t in 2:n_points) {
    theta[t,1] = rnorm(1, rho*theta[t-1,2], v)
    theta[t,2] = rnorm(1, rho*theta[t ,1], v)
  }
  return(theta)
}


theta = gibbs_bivariate_normal(c(-3,3), n<-5000, rho=rho<-0.9)
plot(theta[,1], theta[,2], xlim=c(-3,3), ylim=c(-3,3), pch=19, col = rgb(red = 0, green = 0, blue = 1, alpha = 0.10))

```
